```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatOpenAI(model="gpt-4.1-nano")
result = model.invoke("How many leaves I can take in my comapmny ? I work in meta")
print(result)
```

```python
model = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.1, max_completion_tokens=10)

result = model.invoke("Write a 5 liner poem on cricket")
print(result.content)
```
