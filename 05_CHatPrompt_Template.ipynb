```python
#Simple dynamic prompting example using LangChain's ChatPromptTemplate and ChatOpenAI.
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(model="gpt-4o-mini")

prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple words."
)

chain = prompt | llm | StrOutputParser()

response = chain.invoke({"topic": "RAG in GenAI"})
print(response)
```

```python
#2. Role based prompting example using ChatPromptTemplate.

from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI tutor."),
    ("human", "Explain {topic} in simple terms.")
])

chain = prompt | llm | StrOutputParser()

response = chain.invoke({"topic": "Vector Databases"})
print(response)
```

```python
#3. Multiple variables prompting example using ChatPromptTemplate.

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}."),
    ("human", "Explain {topic} for {audience}.")
])

chain = prompt | llm | StrOutputParser()

response = chain.invoke({
    "role": "data science professor",
    "topic": "RAG",
    "audience": "beginners"
})

print(response)
```

```python
#4. using Chat History in prompts example using ChatPromptTemplate.

from langchain_core.messages import HumanMessage, AIMessage

chat_history = [
    HumanMessage(content="My name is Raj."),
    AIMessage(content="Nice to meet you Raj.")
]

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    ("placeholder", "{chat_history}"),
    ("human", "{question}")
])
```

```python
chain = prompt | llm | StrOutputParser()

response = chain.invoke({
    "chat_history": chat_history,
    "question": "What is my name?"
})

print(response)
```

```python
#5. RAG style prompting example using ChatPromptTemplate.

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(model="gpt-4o-mini")

prompt = ChatPromptTemplate.from_template(
    """
Answer the question only using the context below.

Context:
{context}

Question:
{question}
"""
)

chain = prompt | llm | StrOutputParser()

retrieved_text = "RAG stands for Retrieval Augmented Generation."

response = chain.invoke({
    "context": retrieved_text,
    "question": "What is RAG?"
})

print(response)
```
